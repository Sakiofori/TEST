```python
# Write a Python program to find the title tags from a given html document. 
from bs4 import BeautifulSoup

html_doc = """
<html>
<head>
    <title>This is the title</title>
</head>
<body>
    <p>Some text here</p>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')
title_tag = soup.title

if title_tag is not None:
    print(title_tag.string)
else:
    print("No title tag found")
```

```python
#Write a Python program to extract the text in the first paragraph tag of a given html document. 
from bs4 import BeautifulSoup

# HTML document to be parsed
html_doc = """
<html>
<head>
<title>Example HTML Document</title>
</head>
<body>
<h1>This is a Heading</h1>
<p>This is the first paragraph.</p>
<p>This is the second paragraph.</p>
</body>
</html>
"""

# Create a BeautifulSoup object
soup = BeautifulSoup(html_doc, 'html.parser')

# Find the first paragraph tag
first_paragraph = soup.find('p')

# Extract the text from the first paragraph tag
text = first_paragraph.get_text()

# Print the text
print(text)
```

```python
#Write a Python program to extract all the URLs from the webpage python.org that are nested within <li> tags from  import requests
from bs4 import BeautifulSoup

# Make a request to the webpage
url = "https://www.python.org/"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find all the li tags
li_tags = soup.find_all("li")

# Extract the URLs nested within the li tags
urls = []
for li in li_tags:
    if li.find("a") is not None:
        urls.append(li.find("a")["href"])

# Print the extracted URLs
for url in urls:
    print(url)
```

```python
#Write a Python program to retrieve the HTML code of the title, its text, and the HTML code of its parent.
import requests
from bs4 import BeautifulSoup

# Replace the URL with the webpage you want to retrieve
url = "https://www.amazon.com/"

# Retrieve the webpage
response = requests.get(url)

# Extract the HTML code using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Extract the title, text, and parent HTML code
title = soup.title.string
text = soup.get_text()
parent = soup.parent

# Print the results
print("Title: ", title)
print("Text: ", text)
print("Parent HTML code: ", parent)
```

```python
#Write a Python program to create a Beautiful Soup parse tree into a nicely formatted Unicode string, with a separate line for each HTML/XML tag and string.
from bs4 import BeautifulSoup

# Define an example HTML string to parse
html = '''
<html>
<head>
    <title>Example Page</title>
</head>
<body>
    <h1>Welcome to Example Page</h1>
    <p>This is an example page to demonstrate BeautifulSoup parsing</p>
    <ul>
        <li>Item 1</li>
        <li>Item 2</li>
        <li>Item 3</li>
    </ul>
</body>
</html>
'''

# Create a BeautifulSoup object from the HTML string
soup = BeautifulSoup(html, 'html.parser')

# Define a function to recursively print the parse tree
def print_tree(node, indent=0):
    # Print the current node's tag or string
    if hasattr(node, 'name'):
        print('  ' * indent + '<' + node.name + '>')
    elif isinstance(node, str):
        print('  ' * indent + node.strip())

    # Print the child nodes
    for child in node.contents:
        print_tree(child, indent + 1)

    # Print the closing tag, if applicable
    if hasattr(node, 'name'):
        print('  ' * indent + '</' + node.name + '>')

# Print the parse tree
print_tree(soup)
```

```python
#Write a Python program to remove the contents of a tag in a given html document.
from bs4 import BeautifulSoup

# Sample HTML document
html_doc = """
<html>
<head>
<title>My Webpage</title>
</head>
<body>
<h1>Welcome to my webpage</h1>
<p>This is some text that I want to keep</p>
<div class="remove">This is some text that I want to remove</div>
</body>
</html>
"""

# Create BeautifulSoup object
soup = BeautifulSoup(html_doc, 'html.parser')

# Get the tag to remove
tag_to_remove = soup.find('div', {'class': 'remove'})

# Remove the contents of the tag
tag_to_remove.clear()

# Print the updated HTML document
print(soup.prettify())
```

```python
#Write a Python program to wrap an element in the specified tag and create the new wrapper.
def wrap_element(tag, element):
    """
    Wraps the given element in the specified HTML tag and returns the new wrapper as a string.
    """
    return f"<{tag}>{element}</{tag}>"

# Example usage
element = "Hello, world!"
wrapped_element = wrap_element("div", element)
print(wrapped_element)  # Outputs: <div>Hello, world!</div>
```

```python
#Write a Python program to remove a tag from a given tree of html document and destroy it and its contents.
from bs4 import BeautifulSoup

# Example HTML document
html_doc = """
<html>
<head>
<title>Example HTML Document</title>
</head>
<body>
<div class="content">
<h1>Heading 1</h1>
<p>Paragraph 1</p>
<p>Paragraph 2</p>
</div>
</body>
</html>
"""

# Create a BeautifulSoup object from the HTML document
soup = BeautifulSoup(html_doc, 'html.parser')

# Find the tag to remove (in this example, we'll remove the <div> tag with class="content")
tag_to_remove = soup.find('div', {'class': 'content'})

# Remove the tag and its contents from the document
tag_to_remove.extract()

# Print the modified HTML document
print(soup.prettify())
```
